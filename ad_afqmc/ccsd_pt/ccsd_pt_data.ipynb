```python
import numpy as np
from matplotlib import pyplot as plt
def get_data(data):
    n = []
    e = []
    err = []
    for line in data.splitlines():
        if len(line) > 0:
            m = [int(line.split()[0])]
            n = np.hstack((n,m))
            energy = [float(line.split()[-3])]
            e = np.hstack((e,energy))
            error = [float(line.split()[-1])]
            err = np.hstack((err,error))
    e_per = e/n
    err_per = err/n
    return n,e_per,err_per
```
