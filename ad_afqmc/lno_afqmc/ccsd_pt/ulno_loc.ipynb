```python
from pyscf import gto, scf, mp, cc

# a = 2 # bond length in a cluster
# d = 2.2 # distance between each cluster
# unit = 'b' # unit of length
# na = 2  # size of a cluster (monomer)
# nc = 4 # set as integer multiple of monomers
# spin = 0 # spin per monomer
# frozen = 0 # frozen orbital per monomer
# elmt = 'H'
# basis = '6-31g'
# atoms = ""
# for n in range(nc*na):
#     shift = ((n - n % na) // na) * (d-a)
#     atoms += f"{elmt} {n*a+shift:.5f} 0.00000 0.00000 \n"

atoms = '''
O  -1.551007  -0.114520   0.000000
H  -1.934259   0.762503   0.000000
H  -0.599677   0.040712   0.000000
O   1.350625   0.111469   0.000000
H   1.680398  -0.373741  -0.758561
H   1.680398  -0.373741   0.758561
'''
basis = 'ccpvdz'

mol = gto.M(atom=atoms, basis=basis, spin=0, verbose=4, max_memory=16000)
mf = scf.UHF(mol).density_fit()
mf.kernel()

mo, _ = mf.stability()
dm = mf.make_rdm1(mo,mf.mo_occ)
mf.kernel(dm0=dm)
mo, _ = mf.stability()
dm = mf.make_rdm1(mo,mf.mo_occ)
mf.kernel(dm0=dm)
mo, _ = mf.stability()
```

```python
from pyscf import lo
import numpy as np
nfrozen = 0
orbocca = mf.mo_coeff[0][:,nfrozen:np.count_nonzero(mf.mo_occ[0])]
orboccb = mf.mo_coeff[1][:,nfrozen:np.count_nonzero(mf.mo_occ[1])]

loc_a = lo.PM(mol, orbocca)
orbloca = loc_a.kernel()
stable_a, orbloca = loc_a.stability_jacobi()

loc_b = lo.PM(mol, orboccb)
orblocb = loc_b.kernel()
stable_b, orblocb = loc_b.stability_jacobi()

print(f'local orbital stability: {stable_a}, {stable_b}')

lo_coeff = [orbloca, orblocb]
```

```python
def decomp_t2(t2, cutoff=1e-5):
    '''
    t2 shape (i,a,j,b)
    return t2_left, t2_right
    '''
    ni,na,_,_ = t2.shape
    t2 = t2.reshape(ni*na,ni*na)
    u,s,v = np.linalg.svd(t2)
    t2_l = u.reshape(ni,na,-1)
    t2_l = t2_l.transpose(2,0,1)
    t2_r = v.reshape(-1,ni,na)
    mask_singular = s > cutoff
    s_cut = s[mask_singular]
    t2_l_cut = t2_l[mask_singular,:,:]
    t2_r_cut = t2_r[mask_singular,:,:]
    t2_l_cut = lib.einsum('kia,k->kia', t2_l_cut, np.sqrt(s_cut))
    t2_r_cut = lib.einsum('kia,k->kia', t2_r_cut, np.sqrt(s_cut))
    return t2_l_cut, t2_r_cut
```

```python
t2_test = t2bb.transpose(0,2,1,3)
t2_l_cut, t2_r_cut = decomp_t2(t2_test,cutoff=1e-5)
print(t2_l_cut.shape)
t2_rec = lib.einsum('kia,kjb->iajb',t2_l_cut,t2_r_cut)
print(abs((t2_rec-t2_test)).max())
```
